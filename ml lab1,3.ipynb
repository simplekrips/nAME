```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.random.uniform(-1, 1, 100)
z
```

```python
a = -17
b = 9
```

```python
x = np.random.uniform(-0.5, 0.5, 100)
y = a*x+b+z
plt.scatter(x, y)
```

```python
a1 = 1
b1 = 1
speed = 0.01
```

```python
def J1(a,b,x,y):
    j = 0
    for i in range(100):
        j+=2*(a*x[i]+b-y[i])*x[i]
    return j/(2*100)
```

```python
def J2(a,b,x,y):
    j = 0
    for i in range(100):
        j+=2*(a*x[i]+b-y[i])
    return j/(2*100)
```

```python
for i in range(10000):
    c1 = J1(a1,b1,x,y)
    c2 = J2(a1,b1,x,y)
    a1 = a1 - speed * c1
    b1 = b1 - speed * c2
    #print(c1,c2)
```

```python
plt.scatter(x, y, color = "green")
plt.plot(x,a1*x+b1, color = "red")
```
